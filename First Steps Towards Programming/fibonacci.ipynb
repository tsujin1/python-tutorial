```python
# Fibonacci series:
# the sum of two elements defines the next
a, b = 0, 1
while a < 10:
    print(a)
    a, b = b, a+b
    
# Visualization of each iteration
# Replace a and b with their current values.

# a = 0, b = 1
# 0, 1 = 1, 0 + 1   -> a = 1, b = 1

# a = 1, b = 1
# 1, 1 = 1, 1 + 1   -> a = 1, b = 2

# a = 1, b = 2
# 1, 2 = 2, 1 + 2   -> a = 2, b = 3

# a = 2, b = 3
# 2, 3 = 3, 2 + 3   -> a = 3, b = 5

# a = 3, b = 5
# 3, 5 = 5, 3 + 5   -> a = 5, b = 8

# a = 5, b = 8
# 5, 8 = 8, 5 + 8   -> a = 8, b = 13

# a = 8, b = 13
# 8, 13 = 13, 8 + 13 -> a = 13, b = 21

# Loop stops because:
# while a < 10
# while 13 < 10  -> False
```
